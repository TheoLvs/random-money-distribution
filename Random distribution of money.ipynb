# Object definitions

## Entity

```python
class Entity(object):
    def __init__(self,money = 10):
        self.money = money
        
    def __repr__(self):
        return "Money : {}".format(self.money)
        
        
    def give_money(self,money = 1,random = False):
        if self.money > 0:
            if random:
                m = np.random.randint(self.money)
            else:
                m = 1

            self.money -= m
            return m

        else:
            return 0
        
    def receive_money(self,money):
        self.money += money
```

## Population

```python
len(np.random.randint(0,50,50))
```

```python
class Population(object):
    def __init__(self,money = 10,n = 50):
        self.money = money
        self.entities = [Entity(money = money) for x in range(n)]
        self.n = n
    
    def __repr__(self):
        return " ".join(map(str,self.get_money()))
        
        
    def get_money(self,sort = False):
        if sort:
            return list(sorted(self.get_money(sort = False)))
        else:
            return [entity.money for entity in self.entities]
            
    def show(self):
        plt.figure(figsize = (12,3))
        sns.barplot(x = list(range(self.n)),y = self.get_money(),color="b")
        plt.show()
    
    def give_money(self,money = 1,random = False):
        exchange = [entity.give_money(money = money,random = random) for entity in self.entities]
        pick = np.random.randint(0,self.n,self.n)
        
        for i in range(self.n):
            self.entities[pick[i]].receive_money(exchange[i])
            
            
```

## Animation

```python
import matplotlib.pyplot as plt
from matplotlib import animation

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1,figsize = (15,6))

n = 100
p = Population()
barcollection = ax1.bar(range(p.n),p.get_money())
ax1.set_ylim([0, 50])
barcollection2 = ax2.bar(range(p.n),p.get_money(True),color = "grey")
ax2.set_ylim([0, 50])

def animate(i):
    p.give_money()
    y = p.get_money()
    for i, b in enumerate(barcollection):
        b.set_height(y[i])
        
    y = p.get_money(True)
    for i,b in enumerate(barcollection2):
        b.set_height(y[i])

anim=animation.FuncAnimation(fig,animate,repeat=True,blit=False,frames=n,interval=100)
plt.show()
```
